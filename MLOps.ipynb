```python
# load dataset
from pycaret.datasets import get_data
insurance = get_data('insurance')
type(insurance)

df = insurance[['age','bmi','children','charges']]
```

```python
# init environment
from pycaret.regression import setup,create_model,save_model,get_config

r1 = setup(df, target = 'charges', session_id = 123,
           normalize = True,
           polynomial_features = True)

# train a model
lr = create_model('lr')
```

```python

import numpy as np
import pandas as pd
from pycaret.regression import predict_model

cols =cols = ['age',  'bmi', 'children']
final = np.array(['36', '27.900', '3'])
data_unseen = pd.DataFrame([final], columns = cols)

prediction = predict_model(lr, data=data_unseen, round = 0)
print("***************",prediction)
prediction = int(prediction.prediction_label[0])


```
